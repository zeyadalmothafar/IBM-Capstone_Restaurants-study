```python
My_table = soup.find("table",{"class":"wikitable sortable"})
x=My_table.find_all('td')
```

```python
counter=0
a=[]
b=[]
c=[]
for i in x:
    v=i.next
    if counter==0:
        a.append(v)
        counter+=1
        continue
    if counter==1:
        b.append(v)
        counter+=1
        continue
    if counter==2:
        c.append(v)
        counter=0
```

```python
borough=[]
counter=0
for i in b:
    
    if str(i)!="Not assigned" and str(i)!="York" and str(i)!="Mississauga" :
        counter+=1
        x=str(i).index("title=")+7
        y=str(i).index('">')
        borough.append(str(i)[x:y])
    elif "Scarborough" in str(i):
        borough.append("Scarborough")
    else:
        borough.append(str(i))
        
borough=[i.replace("Scarborough, Toronto","Scarborough") for i in borough]
borough=[i.replace(",Toronto"," Toronto") for i in borough]
borough
print(borough)       
```

```python
postcode=[]
for i in a:
    postcode.append(str(i))
    
print(postcode)
```

```python
neighbourhood=[]
counter=0
for i in c:
    if "</a>"not in str(i):
        neighbourhood.append(str(i))
    else:
        counter+=1
        x=str(i).index("title=")+7
        y=str(i).index('">')
        neighbourhood.append(str(i)[x:y])
        
neighbourhood = [i.replace("\n", "") for i in neighbourhood]   
print(neighbourhood)
```

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        #print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            500, 
            100)
            
        # make the GET request
        #print(requests.get(url).json())
        group_data = requests.get(url).json().get("response").get('groups')
        
        if group_data:
            results = group_data[0].get('items')
        else:
            results=''
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v.get("venue").get("id"),
            v.get('venue').get('name'), 
            v.get('venue').get('location').get('lat'), 
            v.get('venue').get('location').get('lng'),  
            v.get('venue').get('categories')[0].get('name')) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue ID',
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

## Function to get the price, likes count and rating for each restaurant

```python
def getRestaurntDetails(names,IDs):
    
    details_list=[]
    for name,ID in zip(names,IDs):
        #print(name)
            
        # create the API request URL
        url="https://api.foursquare.com/v2/venues/{}?client_id={}&client_secret={}&v={}".format(
            ID,
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION         )
            
        # make the GET request
        
        group_data = requests.get(url).json()
        
        #print(group_data)
         #break
        if group_data.get("response"):
        
            #print(group_data.get("response"))
        
            
            if group_data.get("response").get("venue"):
                if group_data.get("response").get("venue").get("price"):
                    if group_data.get("response").get("venue").get("price").get("message"):
                        
                            details_list.append([(name, 
                                                  group_data.get("response").get('venue').get("price").get("message"),
                                                  group_data.get("response").get('venue').get("rating"),
                                                  group_data.get("response").get('venue').get("likes").get("count")
                                                 ) ])
                            

    
    details = pd.DataFrame([item for detail_list in details_list for item in detail_list])
    #details.head()
    #print(details)
    details.columns = ['Rest_name', 
                       'Price', 
                       'Rating' ,
                       'Likes Count'
                      ]
    
    return(details)
    
```
